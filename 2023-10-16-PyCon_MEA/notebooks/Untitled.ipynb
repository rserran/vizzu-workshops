```python
%%javascript
require.config({paths: {gif: 'https://cdn.jsdelivr.net/npm/gif.js@0.2.0/dist/gif.min'}});
require(['gif'],function(GIF){
    let baseUrl = 'https://cdn.jsdelivr.net/npm/gif.js@0.2.0/dist';
    fetch(baseUrl + '/gif.worker.js')
    .then((response) => {
        if (!response.ok) {
            throw new Error("Network response was not OK");
        }
        return response.blob();
    })
    .then((workerBlob) => {
        window.gif = new GIF({
            workers: 4,
            workerScript: URL.createObjectURL(workerBlob),
            quality: 10,
            width: 640,
            height: 480
        });    
        window.gif.on('finished', function(blob) {
            window.open(URL.createObjectURL(blob));
        });
    })
    .then(() => alert('GIF export initialized'))
});
import pandas as pd
from ipyvizzu import Chart, Data, Config

data_frame = pd.read_csv(
    "https://ipyvizzu.vizzuhq.com/0.15/showcases/titanic/titanic.csv"
)
data = Data()
data.add_data_frame(data_frame)

chart = Chart(width="640px", height="480px")

chart.on("logo-draw", "window.gif.addFrame(event.renderingContext, {copy: true, delay: 25})")

chart.animate(data)

chart.animate(
    Config(
        {
            "x": "Count",
            "y": "Sex",
            "label": "Count",
            "title": "Passengers of the Titanic",
        }
    )
)
chart.animate(
    Config(
        {
            "x": ["Count", "Survived"],
            "label": ["Count", "Survived"],
            "color": "Survived",
        }
    )
)
chart.animate(Config({"x": "Count", "y": ["Sex", "Survived"]}))
%%js
window.gif.render()
```
